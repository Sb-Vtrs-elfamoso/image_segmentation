Functions to use

```python
# Function to compute signed distance from a binary mask
def signed_distance_from_mask(mask):
    Dm = fast_marching(mask)
    Dp = fast_marching(1 - mask)
    phi = Dp - Dm
    return phi

# Function to perform fast marching
def fast_marching(mask):
    nbc, nbl = mask.shape[:2]
    D = np.ones((nbc, nbl)) * 1000000.0

    for i in range(nbc):
        for j in range(nbl):
            if mask[i, j] == 1:
                D[i, j] = 0

    for j in range(nbl):
        for i in range(nbc):
            if i > 0:
                if D[i - 1, j] + 1.0 < D[i, j]:
                    D[i, j] = D[i - 1, j] + 1.0
            if j > 0:
                if D[i, j - 1] + 1.0 < D[i, j]:
                    D[i, j] = D[i, j - 1] + 1.0
            if i > 0 and j > 0:
                if D[i - 1, j - 1] + np.sqrt(2.0) < D[i, j]:
                    D[i, j] = D[i - 1, j - 1] + np.sqrt(2.0)
            if i < nbc - 1 and j > 0:
                if D[i + 1, j - 1] + np.sqrt(2.0) < D[i, j]:
                    D[i, j] = D[i + 1, j - 1] + np.sqrt(2.0)

    for i in range(nbc - 2, -1, -1):
        for j in range(nbl):
            if D[i + 1, j] + 1.0 < D[i, j]:
                D[i, j] = D[i + 1, j] + 1.0

    for j in range(nbl - 1, -1, -1):
        for i in range(nbc - 1, -1, -1):
            if i < nbc - 1:
                if D[i + 1, j] + 1.0 < D[i, j]:
                    D[i, j] = D[i + 1, j] + 1.0
            if j < nbl - 1:
                if D[i, j + 1] + 1.0 < D[i, j]:
                    D[i, j] = D[i, j + 1] + 1.0
            if i < nbc - 1 and j < nbl - 1:
                if D[i + 1, j + 1] + np.sqrt(2.0) < D[i, j]:
                    D[i, j] = D[i + 1, j + 1] + np.sqrt(2.0)
            if i > 0 and j < nbl - 1:
                if D[i - 1, j + 1] + np.sqrt(2.0) < D[i, j]:
                    D[i, j] = D[i - 1, j + 1] + np.sqrt(2.0)

    for i in range(1, nbc):
        if D[i - 1, 0] + 1.0 < D[i, 0]:
            D[i, 0] = D[i - 1, 0] + 1.0

    for j in range(1, nbl):
        if D[0, j - 1] + 1.0 < D[0, j]:
            D[0, j] = D[0, j - 1] + 1.0

    for i in range(nbc - 2, -1, -1):
        if D[i + 1, nbl - 1] + 1.0 < D[i, nbl - 1]:
            D[i, nbl - 1] = D[i + 1, nbl - 1] + 1.0

    for j in range(nbl - 2, -1, -1):
        if D[nbc - 1, j + 1] + 1.0 < D[nbc - 1, j]:
            D[nbc - 1, j] = D[nbc - 1, j + 1] + 1.0

    return D

# Function to update color constants c1 and c2
def update_color_constants(phi, image, eta):
    H_phi = Heaviside(phi, eta)
    c1 = np.sum(image * H_phi) / np.sum(H_phi)
    c2 = np.sum(image * (1 - H_phi)) / np.sum(1 - H_phi)
    return c1, c2

# Function to compute Heaviside function
def Heaviside(phi, eta=1e-5):
    return (1 + 2 * np.arctan(phi/eta)/np.pi)/2

# Function to compute the gradient of an image
def compute_gradient(image):
    grad_x = np.gradient(image, axis=0)
    grad_y = np.gradient(image, axis=1)
    return grad_x, grad_y

# Function to compute the divergence of a gradient
def compute_divergence(grad_x, grad_y):
    div = np.gradient(grad_x, axis=0) + np.gradient(grad_y, axis=1)
    return div

def delta_eta(phi,eta):
    return 1/((1 + (phi/eta)**2)*eta*np.pi)

def grad_J_phi(image, phi, eta, epsilon, lambd, c1, c2):
    grad_x, grad_y = compute_gradient(phi)
    norm = np.sqrt(np.sum(grad_x**2 + grad_y**2 + epsilon**2))
    div = compute_divergence(grad_x, grad_y)
    return -delta_eta(phi, eta) * div/norm + lambd*((image - c2)**2 - (image - c1)**2)

# Function to update phi using gradient descent
def update_phi(image, phi, eta, lambd, c1, c2, epsilon):
    
    grad_J = grad_J_phi(image, phi, eta, epsilon, lambd, c1, c2)
    
    # Compute the updated phi, handling invalid values as well
    tau = 1/(2*np.max(np.max(grad_J)))
    new_phi = phi - tau * grad_J
    return new_phi


# Function to update phi as a signed distance function
def update_phi_signed_distance(phi):
    mask = phi > 0
    new_phi = signed_distance_from_mask(mask)
    return new_phi

# Function to perform level set segmentation
def level_set_segmentation(image, iter_max, mask, eta, lambd, tau, n, threshold, epsilon):
    phi = signed_distance_from_mask(1-mask)
    phi_sequence = [phi.copy()]

    c1, c2 = update_color_constants(phi, image, eta)    

    for iteration in range(iter_max):
        """print(c1, c2)
        plt.imshow(phi)
        plt.colorbar()
        plt.show()
        """
        phi_previous = phi.copy()
        
        phi = update_phi(image, phi, eta, lambd, c1, c2,epsilon)
        
        if iteration % n == 0:
            phi = update_phi_signed_distance(phi)
        
        c1, c2 = update_color_constants(phi, image, eta)

        H_phi_previous = Heaviside(phi_previous, eta)
        H_phi = Heaviside(phi, eta)
        diff = np.linalg.norm(H_phi - H_phi_previous)
        #print(diff)
        phi_sequence.append(phi.copy())
        
        if diff < threshold:
            break

    return phi_sequence
```

```python
# Step 1: Load an image
image = io.imread('eight.png')
image = image.astype(float)  # Convert image to floating-point

# Step 2: Define an initial region of interest (mask)
mask = np.zeros_like(image)
mask[20:220, 20:270] = 1

# Display the image and the mask
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Image')
plt.subplot(1, 2, 2)
plt.imshow(mask, cmap='gray')
plt.title('Initial Mask')
plt.show()
```

```python
# Step 3: Define the parameters for level set segmentation
eta = 0.5
lambd = 0.0001
tau = 0.5
n = 10
threshold = 1
epsilon = 1
iter_max = 300

# Step 4: Perform level set segmentation
phi_sequence = level_set_segmentation(image, iter_max, mask, eta, lambd, tau, n, threshold, epsilon)

# Step 5: Display the transformation of the contour

for i, phi in enumerate(phi_sequence):
    """plt.figure(figsize=(12, 6))
    plt.imshow(image, cmap='gray')
    plt.contour(phi, levels=[0], colors='r', linewidths=2)
    plt.title('Iteration {}'.format(i * n))
    plt.axis('off')
    """

# Step 6: Display the final contour
plt.figure()
plt.imshow(image, cmap='gray')
plt.contour(phi_sequence[-1], levels=[0], colors='r', linewidths=2)
plt.title('Final Contour')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float
from skimage.segmentation import (morphological_chan_vese,
                                  morphological_geodesic_active_contour,
                                  inverse_gaussian_gradient,
                                  checkerboard_level_set)


def store_evolution_in(lst):
    """Returns a callback function to store the evolution of the level sets in
    the given list.
    """

    def _store(x):
        lst.append(np.copy(x))

    return _store


# Morphological ACWE
image = io.imread('eight.png')

# Initial level set
init_ls = checkerboard_level_set(image.shape, 6)
# List with intermediate results for plotting the evolution
evolution = []
callback = store_evolution_in(evolution)
ls = morphological_chan_vese(image, num_iter=50, init_level_set=init_ls,
                             smoothing=3, iter_callback=callback)

fig, axes = plt.subplots(1, 2, figsize=(8, 8))
ax = axes.flatten()

ax[0].imshow(image, cmap="gray")
ax[0].set_axis_off()
ax[0].contour(ls, [0.5], colors='r')
ax[0].set_title("Morphological ACWE segmentation", fontsize=12)

ax[1].imshow(ls, cmap="gray")
ax[1].set_axis_off()
contour = ax[1].contour(evolution[2], [0.5], colors='g')
contour.collections[0].set_label("Iteration 2")
"""contour = ax[1].contour(evolution[7], [0.5], colors='y')
contour.collections[0].set_label("Iteration 7")
contour = ax[1].contour(evolution[-1], [0.5], colors='r')
contour.collections[0].set_label("Iteration 35")
ax[1].legend(loc="upper right")"""
title = "Morphological ACWE evolution"
ax[1].set_title(title, fontsize=12)
```
